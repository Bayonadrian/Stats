# **Binomial Distribution**
It is useful to get the possibility of success within an event, and it is composed by two parts in the same formula: $$Binom = \begin{pmatrix}n \\ k\end{pmatrix} p^k(1-p)^{(n-k)}$$
- n is the total number of elements inside the sample.
- p is the probability of success.
- k is the number in the sample we are looking for.

The first part is: $$\begin{pmatrix}n \\ k\end{pmatrix}$$
which is the same as: $$\frac{n!}{k!(n-k)!}$$

The second part is: $$p^k(1-p)^{(n-k)}$$

So the complete formula is: $$\frac{n!}{k!(n-k)!}(p^k(1-p)^{(n-k)})$$

Just imagine we are the owners of a car service, our average of visits with minor problems is **75%(p)**, let's imagine we will have **60(n)** cars tomorrow, so what is the possibility to have **40(k)** of those 60 with minor problems. So the first part is: 
$$\frac{60!}{40!(60-40)!}=\frac{60!}{40!20!}=4191844505805495.0$$ 

The second part is: 

$$0.75^{40}(0.25)^{(60-40)} = 0.75^{40}*0.25^{20} = 0.0000000009....$$

And putting both together we have:

$$4191844505805495.0*0.0000000009....=0.038340332373061344$$

Finally, the probability to have 40 cars with minor problems is 0.0383.

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

bin = binom(60, 0.75)
prob = bin.pmf(np.arange(0, 61, 1))
print('Distribution:', prob)
moments = {'mean': 0, 'variance': 0, 'skew': 0, 'kurtosis': 0}
moments['mean'], moments['variance'], moments['skew'], moments['kurtosis'] = bin.stats(moments='mvsk')
print('Moments:', moments.items())
print('Probability to get 40 cars with minor problems', bin.pmf(40))
fig, ax = plt.subplots()
ax.bar(np.arange(0, 61, 1), prob, width=0.6)
ax.plot(np.arange(0, 61, 1), prob)
plt.show()
```
