# **Bernoulli Distribution**
It is useful for dichotomic experiments, it has only two possible resultants, which are true(sucess) and false(failure), imagine that we roll a dice and we want to get a 3 so the probability to get a success is: $$\frac{1}{6}=0.16666666666666666
$$
And the oppsite(failure) is: $$\frac{5}{6}=0.8333333333333334$$
Then applying the bernoulli formula we get: $$success^z(1-success)^{1-z}=\frac{1}{6}^z(1-\frac{1}{6})^{1-z}$$
But until this part, what about z?. Z is an especial value, it can be 1 or 0: $$\bigg\{\begin{align*}
1-p => k=0 \\ 
p => k=1
\end{align*}
$$
Therefore, if you want to get the probability of success k=1, otherwise k=0. Back to the initial formula: 
$$\Bigg\{\begin{align*}
\frac{1}{6}^1(1-\frac{1}{6})^{1-1} = \frac{1}{6} \\
\frac{1}{6}^0(1-\frac{1}{6})^{1-0} = \frac{5}{6}
\end{align*}$$

```python
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
import numpy as np

p = 0.6

ber = bernoulli(p)

print('Moments:', ber.stats(moments='mvsk'))
print('PMF:', ber.pmf([1, 0]))

fig, ax = plt.subplots()
ax.bar([1, 0], [ber.pmf(1), ber.pmf(0)], width=0.02)
ax.plot([1, 0], [ber.pmf(1), ber.pmf(0)])
plt.show()
```
