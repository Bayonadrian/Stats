# **Centrality Measures**
Those are ways to get representative values of the data, such as the value of the middle. There are 4 important concepts:
- Mean
- Median
- Mode
- Percentiles

# **Mean**
There is more than one type of mean. To see more info follow this link: https://youtu.be/jXKYI7wyqp0.

## Arithmetic Mean
 It is well known as **"average"** and it is composed by the next formula: $$\frac{\sum{f_{i}}}{N}$$

Where:

 - **"fi"** represent each value in the sample
 - **"N"** is the total quantity of items in the sample

````python
# Imagine we have a list with the age of workers within a company and we want to know the average

import scipy as sp

ages = [22, 24, 22, 23, 26, 25, 24, 27]

sp.mean(ages)
````

## Geometric Mean
It is used to get the mean of changes in data, to make it possible is necessary to use positive and progressive values as an example if something increases one year by 5% and the last year by 15%, the main formula is: $$\sqrt[n]{(x_{1})(x_{2})(x_{3})...(x_{n})}$$

Where:

- **"n"** is the number of values
- **"x"** is the value

---

So to continue with the example the formula is: $$\sqrt[2]{(1.05)(1.15)} = 1.09886$$
In this example we have used 1 because those values were more than a hundred percent, so the correct mean is: $$0.09886$$

````python
import scipy.stats as sp

sp.gmean([1.05, 1.15])
````

## Armonic Mean
It is most commonly used in case of time laps, the furmula is:
$$\frac{n}{\sum{(\frac{1}{x})}}$$

Where:

- N is the number of items in the  sample.
- x is the value of any item.


````python
import scipy.stats as sp

sp.hmean([3,2])
````

## Categorical data Mean
For categorical data there is no mean, it is something really important, but it is possible to find a measure using the proportion wich means the percentage of the value you want to calculate, as an example let's imagine a sample of 300 persons where 120 are males and 180 are females then let's get the proportion for males:

$$proportion = \frac{p}{n}$$

where:

- **p**, is the probability of success.
- **n**, is the sample size.

Then we have:

$$proportion = \frac{120}{300}$$

$$proportion = \frac{120}{300} = 0.4$$




````python
males_proportion = 120 / 300
males_proportion
````

# **Median**
This is the value of the middle in the sample. As an example a sample like that:

```
# [2,3,5,7,8]
```
Has 5 as the value of the middle. However, let's think about what happens with a sample like that:

```
# [1,2,3,4,5,6]
```
This array does not have a specific middle, in this case, it applies this formula:

$$\frac{n_{1}+n_{2}}{N}$$

Where:

- **"n1"** and **"n2"** are the two values in the middle
- **"N"** is the number of values in the middle 

---

The result of the application of the last furmula in this case is: $$\frac{3+4}{2} = \frac{7}{2} = 3.5$$

````python
import numpy as np
np.median([2,3,5,7,8])
````

````python
import numpy as np
np.median([1,2,3,4,5,6])
````

# **Mode**
This is the most common value in a sample.

````python
import scipy.stats as sp

sp.mode([1,2,3,4,5,6,7,3,2,3,2,2,1,1,6,6,4,5,7,3,6,6,6])
````

# **Percentiles**
This is a way to separate the sample in percentages: $$P_{n}$$ represents the n% of the sample, as an example, let's think in this array: 

```
# [10,25,40,60,65]
```

If we look for the value in the 50% position, the result will be 40, lets see that it is also the median, the 50% of the sample is before this value, which means also that the other 50% of the sample is after this value.




````python
import numpy as np

np.percentile([10,25,40,60,65], 50)
````

# **Quartiles**
This is a way to separate the sample into four groups:
- Q1, this is the number in 25% of the sample.
- Q2, this is the number in 50% of the sample.
- Q3, this is the number in 75% of the sample.
- After the group Q3, is the fourth group. 

````python
import numpy as np

data = [10, 15, 20, 15, 20, 30, 35]

print('Q1 ->', np.percentile(data, 25))
print('Q2 ->', np.percentile(data, 50))
print('Q3 ->', np.percentile(data, 75))
````

# Range
The range is the difference between the largest and the smallest value, the result is the spread of the sample.
$$Range= Largest - smallest$$

````python
import numpy as np

# The spread of the data is 25.

data = [10, 15, 20, 15, 20, 30, 35]

_range= max(data) - min(data)
_range, np.ptp(data)
````

# **Interquartile Range**
This is a way to get the range between quartiles.

````python
import scipy.stats as sp

data = [10, 15, 20, 15, 20, 30, 35]

# Range between quartile 2 and 3.

sp.iqr(data, rng= (25, 75))
````

# **Histogram**
A histogram is a very useful graphic to know how is the distribution of a sample. Also, it is very useful to find the central measure. Something very important here is the **bins parameter**, bins means the ranges of the data in the histogram. As an exmaple a bin of 10 means that the data will be grouped 10 by 10.

````python
import matplotlib.pyplot as plt
import numpy as np
import random

data= np.random.normal(50, 14, 3000)

fig, ax= plt.subplots()

ax.hist(data, bins=np.arange(1, 100, 10), rwidth=0.7)

plt.show()
````

# **Outliers using IQR**
Those are values far far away from the median, the formula to get the lower outlier is: $$Q_{1} - (1.5*IQR)$$
Also, the higher outlier has the next formula: $$Q_{3} + (1.5*IQR)$$

---

- Q, means quartile.
- IQR, means **Interquartile range**.


# **Box Plot**
It is an special graphic to see percentiles, it shows the minimum and maximum quantity, also the median and quartiles. It is composed by six parts:

- **Lower whisker**, is the first quartile.
- **Lower quantile**, is the second quartile.
- **Middle**, is the median.
- **Upper quantile**, is the third quartile.
- **Upper whisker**, Is the fourth quartile.
- **Outliers**, This is data out of the scope of the quartiles.

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

data = {'Quartiles': [10, 15, 20, 15, 20, 30, 35], 'With outliers': [10, 15, 20, 15, 20, 30, 35, 35, 66]}

ax.boxplot(data.values())
ax.set_xticklabels(data.keys())
ax.text(1.1, np.percentile(data['Quartiles'], 0), 'Lower whisker')
ax.text(1.1, np.percentile(data['Quartiles'], 25), 'Lower quantile')
ax.text(1.1, np.percentile(data['Quartiles'], 50), 'Median')
ax.text(1.1, np.percentile(data['Quartiles'], 75), 'Upper quantile')
ax.text(1.1, np.percentile(data['Quartiles'], 100), 'Upper whisker')
ax.text(2.05, 66, 'Outlier')

plt.show()
````

# **Z-Values to find outliers**
It is possible to get the z-values to get the outliers. When you get the z-values, z-values mean how many standard deviations the data is from the mean. In addition, it is pretty common to take those intervals:

$$num>3, num<-3$$

The formula to get the z-values is:

$$\frac{num-mean}{stdev}$$

In this way if our number in the sample is 15 and our mean is 45 with an standard deviation of 5 we can use the formula:

$$\frac{15-45}{5} = -6$$

The result is -6 so we can apply the rule:

$$-6 < -3$$

So our value is definitely an outlier.

To conclude, **using z-values is better than using IQR just in case of having a symmetric sample. In case of having a skewed sample, it is highly recommendable to use the IQR method**.

````python
value= 15
mean= 45
stdev= 5

(value - mean)/stdev
````

# ECDF
ECDF is a great graphic to make an exploratory analisis, it has two axis in axis y you can see the proportion and in x axis you have the number of occurrences of the event.

As an example, we play with a fixed dice which more frequently returns 1 and 4, in this way the ECDF graph shows greater proportions there.

````python
#@title Using proportions
import random
import seaborn as sns 
import matplotlib.pyplot as plt

dice = [1, 2, 3, 4, 5, 6]

data= random.choices(dice, weights= [3, 1, 1, 2, 1, 1], k=2000)

sns.ecdfplot(data)
plt.xlabel('Dice results')
plt.show()
````

It is also possible to use counts to make the graph.

````python
#@title Using counts
import random
import seaborn as sns 
import matplotlib.pyplot as plt

dice = [1, 2, 3, 4, 5, 6]

data= random.choices(dice, weights= [3, 1, 1, 2, 1, 1], k=2000)

sns.ecdfplot(data, stat='count')
plt.xlabel('Dice results')
plt.show()
````
