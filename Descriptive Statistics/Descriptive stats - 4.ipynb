# **Covariance**
It is useful to get the relation between two variables, it has this formula: $$cov_{xy} = \frac{\sum{(x_{i} - \overline{x})((y_{i}-\overline{y}))}}{n-1}$$
- **"xi"** is each value in the **"x"** sample.
- **"x"** is the mean in the **"x"** sample.
- **"yi"** is each value in the **"y"** sample.
- **"y"** is the mean in the **"y"** sample.
- If the result is: $$cov>0$$
It means that it has a positive relationship.
- If the result is: $$cov<0$$
It means that it has a negative relationship.
- If the result is: $$cov=0$$
It means that it does not have any relationship.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#xData represents the years of a sample of persons
xData = np.array([10, 12, 20, 40, 50])

#yData represents the money that the persons of the sample earn
yData = np.array([100, 120, 400, 3000, 3500])

data = {'years': xData, 'money': yData}

df = pd.DataFrame(data)
# Both has a positive relationship
print(df.cov())

fig, ax = plt.subplots()

ax.scatter(xData, yData)
ax.plot(xData, yData)
```

# **Correlation**
The correlation was developed to know how strong or weak is the relationship between two variables. It has this formula: $$cor = \frac{cov_{xy}}{\sigma_{x}\sigma_{y}}$$
- **"cov"** is the covariance.
- **"ox"** is the standard deviation of "**x**". 
- **"oy"** is the standard deviation of "**y**".
- A positive result of 1 or near to 1 means that both variables have a positive strong relationship.
- A negative result of 1 or near to 1 means that both variables have a negative strong relationship.
- A result of 0 means that both samples have no relationship.

```python
import pandas as pd
import random
import seaborn as sb

data = {"xData": [], "yData": [], "zData": [], "vData": []}

for i in range(10):

  data['xData'].append(random.random())
  data['yData'].append(random.random())
  data['zData'].append(random.random())
  data['vData'].append(random.random())

df = pd.DataFrame(data)
print(df.corr())
sb.heatmap(df.corr())
```

## Correlation Direction

- Positive, A positive correlation is when variables increase together. Then, there is a relationship between variables.
- Negative, A negative correlation is when variables decrease together. Then, there is a relationship between variables.
- No correlation, when is no pattern, there is no relationship between variables.

### Positive

```python
import pandas as pd
import matplotlib.pyplot as plt

data= {'Years studying': [year for year in range(8)], 'Wage': [wage for wage in range(0, 800, 100)]}

df= pd.DataFrame(data)
df
```

```python
df.corr()
```

```python
plt.scatter(data['Years studying'], data['Wage'])
plt.show()
```

### Negative

```python
import pandas as pd
import matplotlib.pyplot as plt

data= {'years': [year for year in range(8)], 'resale price': [wage for wage in range(800, 0, -100)]}

df= pd.DataFrame(data)
df
```

```python
df.corr()
```

```python
plt.scatter(df['years'], df['resale price'])
plt.show()
```

### No correlation

```python
import pandas as pd
import random
import matplotlib.pyplot as plt

data= {'Time playing videogames': [], 'Wage': []}

for _ in range(50):

  data['Time playing videogames'].append(random.randint(1, 1000))

for _ in range(50):

  data['Wage'].append(random.randint(1, 1000))

df= pd.DataFrame(data)
df.head()
```

## Correlation Form

- Linear, It is represented as a linear curve, there is an straight line relationship.
- Non linear, It is represented as a gentle curve, there is non linear relationship.

## Correlation Strength

- Strong correlation, points are very close forming a line or curve.

$$\text{between 0.5 to 1}$$

- Moderate correlation, points are around a generally straight line.

$$\text{0.5}$$

- Weak correlation, points are dispersed but there is a noticeable pattern.

$$\text{between 0 to 0.5}$$

- No correlation, No pattern in the data.

$$\text{0}$$
