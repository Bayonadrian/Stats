# Diseño experimental

La unica forma de probar una relacion de causa efecto es mediante experimentos, los cuales poseen multiples componentes, tales como:

1. Factores, son los atributos que son manipulados.
2. Niveles, son los valores que se le pueden asignar a los factores.
3. Tratamiento, la combinacion de factores y niveles es el tratamiento.
4. Respuesta, es la variable medida.
5. Sujeto, es el sujeto al cual se le aplica el experimento.

Como por ejemplo si se quiere hacer un experimento para definir en que punto las dosis de cierto medicamento es efectiva se tiene la dosis(factor), los niveles(miligramos, como 0mg, 50mg, 100mg), si contrastamos el respuesta(respuesta) de nivel 0mg con la de 50mg, la de 50mg con la de 100mg y la de 0mg con la de 100mg se obtiene el tratamiento(tratamiento). Algo importante hasta este punto es que los niveles se deben probar con sujetos(sujeto), estos son los objetos de los experimentos.

## Principios del diseño experimental

Los cuatro principios del diseño experimental son:

- Aleatorizar: Asignar tratamientos a sujetos al azar, La aleatorización nos permite igualar los efectos de fuentes de variación desconocidas o incontrolables.
La aleatorización hace posible el uso de métodos de inferencia.
La aleatorización brinda protección contra efectos que no puede predecir
“Controla lo que puedas, aleatoriza el resto”.

- Control: Haga que los factores que no estamos probando sean lo más similares posible para todos los grupos de tratamiento, Control. Haga que los factores que no estamos probando sean lo más similares posible para todos los grupos de tratamiento.
Idealmente, la única diferencia entre los grupos debería ser la manipulación
Esto hace que sea más fácil discernir las diferencias entre los tratamientos.

  (Hay un segundo significado de control. Un tratamiento de control es una clase de tratamiento especial diseñada para marcar la línea de base del estudio, y el grupo que lo recibe se denomina grupo de control).

  - Cegamiento: Ocultar deliberadamente los detalles del tratamiento a las personas que podrían afectar el resultado.
Dos fuentes de sesgo no deseado:
Quienes puedan influir en los resultados (los sujetos, administradores del tratamiento, técnicos, etc.)
Quienes evalúan los resultados (jueces, experimentadores, etc.)

  - Placebo: es un tratamiento falso que se parece al tratamiento que se está probando.

- Repetir: Repetir las observaciones para cada tratamiento.
Hacer más de una observación en cada nivel nos permite estimar la variabilidad de las medidas.
Si el número de repeticiones es el mismo para cada combinación de tratamientos, se dice que el experimento está balanceado.

  Un segundo tipo de replicación consiste en repetir un experimento completo para un grupo diferente de sujetos, en circunstancias diferentes o en un momento diferente.
- Bloque: Agrupar sujetos experimentales que sean similares entre sí.
  Los factores de bloqueo se forman de acuerdo con algún factor que no puede controlar y siente que puede afectar la respuesta.

  Los niveles de los factores de bloqueo se denominan bloques.
Ejemplo de factores de bloqueo: sexo, etnia, estado civil, etc.

  En efecto, dividir un experimento en n bloques equivale a ejecutar n experimentos separados.

## Formas de diseño experimental

Existen tres formas de diseño experimental, estas son:

1. Diseños completamente aleatorios.

  - Cuando cada uno de los posibles tratamientos se asigna al menos a un sujeto al azar, el diseño se denomina diseño completamente al azar.

  - En el diseño completamente aleatorizado más simple, los sujetos se asignan aleatoriamente a dos tratamientos.

2. Diseños de bloques aleatorios.

  - Cuando tenemos un factor de bloqueo, aleatorizamos el sujeto a los tratamientos dentro de cada bloque.

3. Diseños Factoriales.

  - Un experimento con más de un factor manipulado se llama diseño factorial.
  - Un diseño factorial completo contiene tratamientos que representan todas las combinaciones posibles de todos los niveles de todos los factores.
  - Cuando la combinación de dos factores tiene un efecto diferente al que se esperaría al sumar los efectos de los dos factores, ese fenómeno se denomina interacción.
  - A menos que su experimento incorpore un diseño factorial, no puede ver las interacciones y esto puede ser una omisión grave.

# Anova(F-statistic)
De momento tenemos herramientas para poder realizar pruebas de hipotesis con una y dos muestras, pero que pasa si es que tenemos mas de dos muestras, aca es donde podriamos usar la prueba t-student multiples veces, pero aca se presenta un problema y este es que al emplear la prueba t-student multiples veces la probabilidad de error tambien aumenta, como por ejemplo si se tiene una probablidad de exito de 95% y se realiza la prueba 4 veces esta probabilidad, se tiene que:

$$0.95^4 = 0.8145$$

Entonces la probabilidad de exito se reduce a poco mas de 81%. Para poder evitar esto y hacer pruebas multiples se puede emplear **Anova(Analysis of variance)**.

Existen dos tipos principales de Anova los cuales son:

- One way(Cuando se tiene un factor).

- Two way(Cuando se tienen dos factores).

- Multiple Anova(Cuando se tienen mas de dos factores).

# One Way Anova

## Supuestos

- Supuesto de independencia: las muestras deben ser independientes entre sí.

- Supuesto de varianza igual: las varianzas de las poblaciones de las que se extrajeron las muestras son iguales. Es decir, $\sigma^2_1 = \sigma^2_2 = \sigma^2_3...$.

- Condición de poblacion normal: La composicion de las muestras debe ser normal.

## Hipotesis

$H_0$ (Hipotesis nula caso one way anova) | $H_a$ (Hipotesis alterna caso one way anova)
--- | ---
$H_0: \mu_1 = \mu_2 = ... \mu_k$ | $H_a: \text{La menos una de las medias es diferente}$ 

## Formula

Existe una formula para poder obtener el estadistico **F**, esta formula tiene la siguiente forma:

$$F = \frac{MS_B}{MS_W}$$

$$MS_B = \frac{\text{Sum of square between sample}}{K - 1}$$

$$MS_W = \frac{\text{Sum of square within sample}}{n_T - k}$$

### Data

Se trabajara con data imaginaria simulando las pruebas de cierto farmaco donde se tienen 3 niveles(placebo, high, low) y un solo factor(dosaje).

```python
import numpy as np

placebo= [3,2,1,1,4]
low= [5,2,4,2,3]
high=	[7,4,5,3,6]
total= placebo + low + high
```

### $MS_B$

#### Between samples 

$$\text{Sum of squares between samples} = \sum\text{Sample size}(\text{mean sample - mean total})^2$$

```python
# Means

mplacebo= np.mean(placebo)
mlow= np.mean(low)
mhigh= np.mean(high)
mtotal= np.mean(total)

mplacebo, mlow, mhigh, mtotal
```

```python
# Sum of squares between samples

sumOfSquaresBetween = 5*(mplacebo-mtotal)**2 + 5*(mlow-mtotal)**2 + 5*(mhigh-mtotal)**2
sumOfSquaresBetween
```

#### K

$$k = \text{Num of samples}$$

```python
# K

k = 3 # placebo, low, high
```

```python
msb = sumOfSquaresBetween / (k-1)
msb
```

### $MS_W$

#### Within samples

$$\text{Sum of squares within samples} = \sum(\text{Sample size} - 1)\text{variance sample}$$

```python
# Variances

vplacebo= np.var(placebo, ddof=1)
vlow= np.var(low, ddof=1)
vhigh= np.var(high, ddof=1)
vtotal= np.var(total, ddof=1)

vplacebo, vlow, vhigh, vtotal
```

```python
# Sum of squares within samples

sumOfSquaresWithin =  ((len(placebo)-1) * vplacebo) + ((len(low)-1) * vlow) + ((len(high)-1) * vhigh)
sumOfSquaresWithin
```

#### K

$$k = \text{Num of samples}$$

```python
# nT -k

len(total) - k
```

```python
msw = sumOfSquaresWithin / (len(total) - k)
msw
```

### F-Statistic

$$F = \frac{MS_B}{MS_W}$$

```python
f = msb/msw
f
```

## Empleando statsmodels

### Convirtiendo la data a DataFrame
Para poder usar estatsmodels es necesario emplear la libreria pandas y convertir la data en un DataFrame.

```python
import pandas as pd

df = pd.DataFrame({
    'Placebo': placebo,
    'High': high,
    'Low': low
})

df
```

### Melting

Para poder realizar un One Way Anova Test, es neceario juntar toda la data y esto se realiza mediante un proceso llamado melting.

```python
df_melted = pd.melt(df, value_vars= ['Placebo', 'High', 'Low'])
df_melted
```

### Anova

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('value ~ variable', data= df_melted).fit()
anova_model = sm.stats.anova_lm(model, typ= 2)
anova_model
```

## Tukey one way anova test
Si bien es cierto que es posible saber si una muestra es diferente al resto empleando anova la informacion no es completa, ya que falta saber cual de las muestras es diferente y en que medida, aqui es donde entra el test de Tukey.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

tukey = pairwise_tukeyhsd(df_melted['value'], groups= df_melted['variable'], alpha= 0.05)
tukey.plot_simultaneous()
tukey.summary()
```

# Two Way Anova

Two Way Anova es para poder trabajar con dos factores, es esta la diferencia con One Way Anova.

## Supuestos

- Supuesto de independencia: las muestras deben ser independientes entre sí.

- Supuesto de varianza igual: las varianzas de las poblaciones de las que se extrajeron las muestras son iguales. Es decir, $\sigma^2_1 = \sigma^2_2 = \sigma^2_3...$.

- Condición de poblacion normal: La composicion de las muestras debe ser normal.

## Hipotesis 

$H_0$ (Hipotesis nula caso two way anova) | $H_a$ (Hipotesis alterna caso two way anova)
--- | ---
Factor A: Las medias de los niveles del factor A son iguales | Al menos una de las medias es diferente a las otras
Factor B: Las medias de los niveles del factor B son iguales |Al menos una de las medias es diferente a las otras 
Interaccion: El efecto del factor A es constante a traves de los niveles del factor B |Al menos una de las medias es diferente a las otras 
 

## Data

La siguiente data sera de un experimento de coles, donde se emplearan dos factores, los cuales son:

- Salinidad.
- Temperatura.

Estos factores se emplearan para determinar la Biomasa de las coles.

```python
data = {'Biomass': [12.27, 12.8, 10.43, 11.8, 11.89, 10.56, 10.02, 7.5, 10.68, 7.66, 8.6, 8.57, 12.13, 12.36, 10.48, 10.72,
                    10.64, 10, 10.62, 8.94, 9.43, 9.66, 8.99, 8.36, 8.24, 9.48, 8.61, 8.96, 8.27, 8.63, 8.49, 8.04, 7.17,
                    7.44, 7.17, 7.06],
        'Salinity': [0, 0, 0, 4, 4, 4, 8, 8, 8, 12, 12, 12, 0, 0, 0, 4, 4, 4, 8, 8, 8, 12, 12, 12, 0, 0, 0, 4, 4, 4, 8, 8,
                    8, 12, 12, 12],
        'Temperature': [36, 36, 36, 36, 36, 36, 36, 36, 36, 36, 36, 36, 34, 34, 34, 34, 34, 34, 34, 34, 34, 34, 34, 34, 32,
                        32, 32, 32, 32, 32, 32, 32, 32, 32, 32, 32]
}
```

```python
import pandas as pd

sprouts_df = pd.DataFrame(data)
sprouts_df.head()
```

## Anova Test

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('Biomass ~ C(Salinity) + C(Temperature) + C(Salinity):C(Temperature)', data= sprouts_df).fit()
anova_model = sm.stats.anova_lm(model, typ= 2)
anova_model
```

## Tukey Test

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Primeramente hay que convertir la data a str
sprouts_df = sprouts_df.astype({'Temperature': 'str'})

sprouts_df = sprouts_df.astype({'Salinity': 'str'})

# Es necesario juntar los factores en una sola columna poniendo el simbolo / en el medio
sprouts_df['Combination'] = sprouts_df['Salinity'] + " / " + sprouts_df['Temperature']

tukey = pairwise_tukeyhsd(sprouts_df['Biomass'], groups= sprouts_df['Combination'], alpha= 0.05)
tukey.plot_simultaneous()
tukey.summary()
```
