# Sampling
Sampling is the easy way to make statistical analisis using just a portion of the total data instead of using the whole population. Something really important about the sample is that every sample has to be taken randomly, also it has to have the correct size to be representative.

## Types of samples
There is a lot of ways to take a sample such as:

- Simple Random Sample.
- Stratified Random Sample.
- Cluster Sample.
- Systematic Sample.
- Multistage Sample.


### Simple Random Sample
The sample has to be taken randomly.

```python
import numpy as np

data= np.arange(1, 100, 1)

sample= np.random.choice(data, 10)

# Sample of 10 itmes from the sample.
sample
```

### Sistematic sample
To get this sample you have to make intervals which are gonna be useful to get the data, to get this sample you have to use the **population size** and the **sample size**.

```python
import numpy as np

population_size = 3000
sample_size = 60

np.random.seed(61)

population = np.random.normal(loc= 50, scale= 24, size= population_size)

interval = population_size // sample_size

sample = list()

# Counter has to be 0 to start from the first one and 1 to start from the first 50 items.

counter = 0

for i in enumerate(population):

  if counter % interval == 0:

    sample.append(i)

  counter += 1


print('Interval: ', interval)

# We take numbers every 50 items within the list(Take the sample with laps of 50 items).

print('Sample: ', sample)
```

### Stratified Random Sample
It is very useful when data can be broken into distincs sub groups, as an example just imagine you want to make an study about people above 1.70 of height so you can broke this data into 3 groups by sex, such as:

- Male.
- Female.
- Not binary.

Stratas has to be defined by the whole population; as an example imagine you have a **population** of 1000 persons and also you know that there is 60% of males and 40% of females and you have to take a sample of 500 persons within the population, then 60% of this sample has to be of males and 40% has to be of females.


```python
import random

population= [[],[]]

for _ in range(1, 601, 1):

  population[0].append('male')

_ = 0

for _ in range(1, 401, 1):

  population[1].append('female')

_ = 0

print('Percentage of males:', len(population[0])/(len(population[0]) + len(population[1])))
print('Percentage in females:', len(population[1])/(len(population[0]) + len(population[1])))

sample_size = 500

sample = list()

# Sample size has to be multiplied by 60% for males following the original distribution.
for s1 in range(int(sample_size*0.6)):

  # Sample subjects within the population has to be selected randomly.
  sample.append(random.choice(population[0]))

# Sample size has to be multiplied by 40% for females following the original distribution.
for s1 in range(int(sample_size*0.4)):

  # Sample subjects within the population has to be selected randomly.
  sample.append(random.choice(population[1]))

# 60% are males and 40% are females in the new sample.
print('Sample size:', len(sample))
```

# Relative errors of point stimates

## Point stimates
Point stimates means that as greatest the sample the closest to the mean of the population.

```python
import numpy as np
import random

np.random.seed(45)
random.seed(99)

# The population size is 5000, loc is the mean and scale is the standard deviation.
population = np.random.normal(loc= 50, scale= 42, size= 5000)

print('mean:', population.mean())

fs = random.choices(population, k= 500)

print('mean first sample:', np.mean(fs))

ss = random.choices(population, k= 1000)

print('mean second sample:', np.mean(ss))

ts = random.choices(population, k= 2000)

print('mean third sample:', np.mean(ts))
```

## Relative error
The relative error is the percentage difference between the mean of the population and the sample mean.

```python
first = (abs(population.mean() - np.mean(fs)) / population.mean()) * 100
second = (abs(population.mean() - np.mean(ss)) / population.mean()) * 100
third = (abs(population.mean() - np.mean(ts)) / population.mean()) * 100

first, second, third
```

```python
import matplotlib.pyplot as plt

plt.bar(['First', 'Second', 'Third'], [first, second, third])
plt.title('Evolution in percentage')
plt.show()
```

# Defining sample size
Something really important about sampling is the sample size because as larger the size, the better the result.

## Formula
To define the sample size is posible to apply the next formula:

$$n = (\frac{t_{n-1} \cdot{sd}}{ME})$$

$n$ = Sample size.

$t_{n-1}$ = The critical value.

$sd$ = Standard Deviation.

$ME$ = Margin error.

### Problems with the formula
Now is when we can realize about some problems using the formula:

|Value|Problem|Solution|
|---|---|---|
|sd|We do not know the standard deviation because we are using a sample|If you have a guess as an example a prior study sd is a good start point or you can make a pilot|
|$t_{n-1}$|We are trying to find n but we have to use n in the formula|For samples of 60 or more we can use z instead. Otherwise, if sample is less tha 60 we can use z to find n

We can find n using z with the next procedure:

$$n = (\frac{z\cdot{sd}}{ME})$$

Then replacing values we can get the value of n and then apply the correct formula with the new n value:

$$n = (\frac{t_{n-1}\cdot{sd}}{ME})$$

### Example
In the 150 samples of farmed salmon the mean concentration of Mirex was 0.0913 ppm with a standard deviation of 0.0495 ppm. How large a sample would be needed to produce a 95% confidence interval with a margin of error of 0.004 in a new study for another farmed salmon?

```python
#@title Getting n
import scipy.stats as st

# Getting z with 95% confidence interval.
z = st.norm.ppf(0.975)
# Standard deviation of another study.
sd =  0.0495
me = 0.0004
# First n.
n = (z * sd)/me

n = round(n)

# Getting t.
t = st.t.ppf(0.975, n-1)

# Getting n.
n = (t * sd)/me

n
```

## Power analysis
Power analisis is a way to define the sample size using a simulation.

### Example

Suppose you are in charge of a media website and you are interested in increasing the amount of time users spend on your website. Currently, the time that users spend on your website is normally distributed with a mean of 1 minute and a standard deviation of 0.5 minutes. Suppose you're introducing a feature that loads pages faster, and you want to know the sample size needed to measure a 5% increase in time spent on the website.

In order to make this power analysis, we are going to propose that the data be distributed using a t-student distribution.

```python
import numpy as np
import scipy.stats as st

# Defining initial statistics.
increment_size = 0.05 
sample_size = 50
initial_distribution_mean = 1
initial_distribution__sd = 0.5

# Defining the new distribution mean.
new_distribution_mean = initial_distribution_mean*(1+increment_size)

# Getting the initial distribution value.
initial_distribution = np.random.normal(loc=initial_distribution_mean, 
                                        scale=initial_distribution__sd, 
                                        size=sample_size)
# Getting the new distribution value.
new_distribution = np.random.normal(loc=new_distribution_mean, 
                                    scale=initial_distribution__sd, 
                                    size=sample_size)
# Getting the two-ttest p-value.
t_stat, p_value = st.ttest_ind(new_distribution, initial_distribution)

# Getting if the p-value is less than 0.05(level of significance).
stat_sig = p_value < 0.05

print("P-value: {}, Statistically Significant? {}".format(p_value, stat_sig))
```

This firts analysis shows that the data is not significant in a lot of instances, then we can use the same logic in order to create the simulation till get an 80% of power, which means at least an 80% of the p-values are less than 0.05(level of significance) having 1000 simulations.

Then we have to get the 80% of p-values less than 0.05 of the 1000 simulations.

```python
sample_size = 50
sims = 1000

while True:
    initial_distribution = np.random.normal(loc=initial_distribution_mean, 
                                            scale= initial_distribution__sd, 
                                            size=(sample_size, sims))
    new_distribution = np.random.normal(loc=initial_distribution_mean*(1+increment_size),
                                        scale= initial_distribution__sd,
                                        size=(sample_size, sims))
    
    # Getting the two-ttest p-value

    t, p = st.ttest_ind(new_distribution, initial_distribution)
    
    # Getting the values where p-value less than 0.05, sum the quantity and get the percentage against the simulations.
    power = (p < 0.05).sum()/sims

    if power >= 0.8: 

        break

    else: 
        # If the percentage of p-values is less than 80% increment 10 to the sample size.
        sample_size += 10

print("For 80% power, sample size required = {}".format(sample_size))
```
